# WHAT IS MARKDOWN?
---

- **Markdown** is a lightweight markup language that you can use to add formatting elements to plaintext text documents. Created by **John Gruber** in 2004, Markdown is now one of the world’s most popular markup languages.<br>

    If you want to learn more about markdown, check out this website - www.markdownguide.org<br><br>

- Markdown can be used for everything. People use it to create websites, documents, notes, books, presentations, email messages, and technical documentation.

- Markdown is portable. Files containing Markdown-formatted text can be opened using virtually any application. If you decide you don’t like the Markdown application you’re currently using, you can import your Markdown files into another Markdown application. That’s in stark contrast to word processing applications like Microsoft Word that lock your content into a proprietary file format.

- Markdown is everywhere. Websites like Reddit and GitHub support Markdown, and lots of desktop and web-based applications support it.

# HOW TO INSERT AN IMAGE USING MARKDOWN:
---

## Inserting an Image saved on your PC:

### Syntax: 
   
```markdown
![image name](full image address on your computer)
``` 

> **Note:**This code **does not work in Jupyter Notebook** but works well in ***Obsidian Application***.
   
## Inserting an Image from Web:

### Syntax:
   
```markdown
![image-name](image url)
```

**Example Code:** `![Steve-Jobs](https://addicted2success.com/wp-content/uploads/2017/11/10-Things-We-Can-Learn-From-the-Incredible-Steve-Jobs.jpg "An example of Genius with a flavour of an Ambitituos mind")`

## Inserting an Image in Jupyter Notebook:

1. Go to **Edit** --> **Insert Image** ---> **Choose Image**.
2. A code will be added in place of the image.
3. Run the code in the markdown cell to get the image printed.

# LINE BRAEKS IN MARKDOWN:
---

Sometimes, Markdown does not add line breaks in your text when you want them. To force add a line break use $<br>$ command at the end of the line.

> **Example:**<br>
> **Code:**
```markdown
> When I say 'Leave Me':<br>
> Just leave me!
```
>**Output:**<br>

> When I say 'Leave Me':<br>
> Just leave me!

# ADDING COLOURED NOTE BOXES:
---

Use one of the $<div>$ tags to display text in a coloured box.<br>

<div class = "alert alert-block alert-info">
    <b>Special Note:</b> Not all Markdown code displays correctly within $<div>$ tags, so review your coloured boxes carefully.
        For example, to make a word bold, surround it with HTML code for bold (<b>text</b>) instead of the Markdown code.
</div>

The colour of the box is determined by the alert type you specifiy.<br>

- Blue boxes (alert-info)

```markdown
<div class="alert alert-block alert-info">
<b>Tip:</b> Use blue boxes (alert-info) for tips and notes. 
If it’s a note, you don’t have to include the word “Note”.
</div>
```

- Yellow boxes (alert-warning)

```markdown
<div class="alert alert-block alert-warning">
<b>Example:</b> Use yellow boxes for examples that are not 
inside code cells, or use for mathematical formulas if needed.
</div>
```

- Green boxes (alert-success)

```markdown
<div class="alert alert-block alert-success">
<b>Up to you:</b> Use green boxes sparingly, and only for some specific 
purpose that the other boxes can't cover. For example, if you have a lot 
of related content to link to, maybe you decide to use green boxes for 
related links from each section of a notebook.
</div>
```

- Red boxes (alert-danger)

```markdown
<div class="alert alert-block alert-danger">
<b>Just don't:</b> In general, avoid the red boxes. These should only be
used for actions that might cause data loss or another major issue.
</div>
```
<div class = "alert alert-block alert-success">
    <b>NOTE:</b> This syntax is not supported in <b>Obsidian</b>.
</div>

# INTERNAL HYPERLINKS:
---

1. To define the destination (i.e., the section on the page or the cell you want to jump to), you just need to insert an empty HTML anchor tag and give it an id.<br>
Insert this code just above **the destination** of your Markdown text that you wanna go to - when you click this link. 
```markdown
<a id='the-destination'></a>
```
You can type the **name** of **your own destination** that you wanna go to when you click this link, in place of $the-destination$ in the code.<br>
The any spaces and special symbols in the **destination name** should be should be replaced by a **hyphen** (-) when creating its **id tag name**.<br><br>

1. Now we have to create the hyperlink to the `<a id='the_destination'></a>` anchor tag that we just created.
We can just use the Markdown syntax:
`[Link to the destination](#the_destination)`

# CREATING SPACES:
---

1. **Use `&nbsp;` infront of your code for single space.**<br>
1. **Use `&ensp;` infront of your code for 2 spaces.**<br>
1. **Use `&emsp;` infront of your code for 4 spaces.**

**Use Case Example:**

**Input:**

```markdown
Hello<br>
&emsp;I am a man.<br>
&emsp;&emsp;I am a man.<br>
&emsp;&emsp;&emsp;&emsp;I am a man.<br>
&nbsp;&nbsp;&nbsp;&nbsp;I am a man.
```

Hello<br>
&emsp;I am a man.<br>
&emsp;&emsp;I am a man.<br>
&emsp;&emsp;&emsp;&emsp;I am a man.<br>
&nbsp;&nbsp;&nbsp;&nbsp;I am a man.

# ADDING CODES IN MARKDOWN:
---

1. To add an **inline code** in Markdown: Insert 1 backtick $(`)$ (that is on the top left corner of your key board just below *esc*) on each side of your code.

   **Example:**
   
   **Code:**
   
   ```markdown
   This is an inline code - `print("Hello World")`
   ```
   
   **Output:**
   
   This is an inline code - `print("Hello World")`
   
   
2. To add a **standalone code**, use the following syntax:

 
   $```(language-name)$

   your_code 
 
   $```$

## Alignment

*You can align text in the columns to the left, right, or center by adding a colon (:) to the left, right, or on both side of the hyphens*

**Code:**
```markdown
| Column 1 | Column 2 | Column 3 | Column 4 |
| :---- | :---------------- | :-----------:  | ----------: |
| Pritam | 24 | Btech | Mechanical Engineering |
| Pradhugyna | 23 | Btech | Mechanical Engineering |
```
**Output:**

| Column 1 | Column 2 | Column 3 | Column 4 |
| :---- | :---------------- | :-----------: | ----------: |
| Pritam | 24 | Btech | Mechanical Engineering |
| Pradhugyna | 23 | Btech | Mechanical Engineering |
